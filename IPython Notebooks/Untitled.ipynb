# Importing / loading your data

First, you should load your data. I will use the word "load" here so there is no confusion about importing versus loading. Importing in Python refers exclusively to putting the contents of a package into your workspace. If you try to "import" your data, it will not work. Instead, we import a package that will allow us to load our data. 

The data will be working with is behavioral data from Christine. Below, we import a package called pandas and use it to load an Excel spreadsheet. Pandas is a useful package for managing your data 

```python
import pandas as pd

filename = r'../Data/EtOH WD Study Combined Raw Data_WM.xlsx'
df = pd.read_excel(filename)

df #Displays a truncated DataFrame. 
```

A DataFrame is basically an array in the sense that it has columns and rows. However, while arrays don't typically have labels associated with its columns and rows, a DataFrame does. Also DataFrames can be filled with mixed data types (i.e., strings and numbers can coexist in the same DataFrame, but this is not commonly done in arrays). Let's look at all the labels of our DataFrame.

```python
# Lists the names of all the columns. 
df.columns

# Here, df is the DataFrame and columns is an attribute of this particular DataFrame. Think of "columns" as an adjective that 
# describes a property of df. 
```

The first four columns are fairly intuitive. The rest are labels for different fear conditioning exposures. BL_A refers to baseline freezing in context A, Tr1_A refers to the first training session context A, E1 to the first extinction session, etc. There are also some unnamed columns that are empty. 

# Tidying your data

First, let's clean up our data. We will only be working on data from the "Normative" experiment. Use a Boolean vector to only select the rows of the DataFrame where the experiment is "Normative". 

```python
# Recall that = sets a variable while == queries whether the statement is True. 
good_rows = df['Experiment'] == 'Normative'

# Using brackets [] (NOT PARENTHESES) following a DataFrame will select the columns of that DataFrame. Here, we look at the 
# "Experiment" column.
# The == queries whether df['Experiment'] is 'Normative' for each row. 

# The result is a Boolean vector that is True for all rows where the Experiment is Normative, and False otherwise. 
good_rows
```

What does the DataFrame look like after it's been filtered?

```python
# loc[] allows you to select certain rows in your DataFrame. 
# Recall that df.columns displayed all the labels of our Excel spreadsheet. loc[] here behaves in a similar way. While columns
# could be thought of as an adjective describing a property of df, loc[] works more like a function (i.e., a verb). Here, we are
# asking Python to select only the locations indicated by good_rows (i.e., where good_rows is True). 
normative = df.loc[good_rows]
normative # Now we only have 
```
